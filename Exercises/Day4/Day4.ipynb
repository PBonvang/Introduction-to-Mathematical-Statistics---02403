```r
hist(Y, col = "blue", nclass = 30)
```

# E2.16

## a)

```r
n1 = 5
n2 = 8
mu1 = 2
mu2 = 4

sigma = sqrt(2)

k = 10000

S1 <- replicate(k, var(rnorm(n1, mean = mu1,  sd = sigma)))
S2 <- replicate(k, var(rnorm(n2, mean = mu2,  sd = sigma)))

Sp <- ((n1-1)*S1 + (n2-1)*S2)/(n1+n2-2)
Y <- (n1+n2-2)/(sigma^2) * Sp
```

```r
par(mfrow = c(1, 2))

hist(Y, freq = FALSE)
curve(dchisq(xseq, df = n1+n2-2), xname="xseq", add =TRUE, col = 2)

plot(ecdf(Y))
curve(pchisq(xseq, df = n1+n2-2), xname="xseq", add =TRUE, col = 2)
```

## b)

```r
n1 = 5
n2 = 8
mu1 = 2
mu2 = 4

sigma = sqrt(2)

k = 10000

X_mean = replicate(k, mean(rnorm(n1, mu1, sigma)))
Y_mean = replicate(k, mean(rnorm(n2, mu2, sigma)))

S1 <- replicate(k, var(rnorm(n1, mean = mu1,  sd = sigma)))
S2 <- replicate(k, var(rnorm(n2, mean = mu2,  sd = sigma)))

Sp <- ((n1-1)*S1 + (n2-1)*S2)/(n1+n2-2)

Y <- (X_mean-Y_mean-(mu1-mu2))/(sqrt(Sp*(1/n1 + 1/n2)))
```

```r
par(mfrow = c(1, 2))

hist(Y, freq = FALSE)
curve(dt(xseq, df = n1+n2-2), xname="xseq", add =TRUE, col = 2)

plot(ecdf(Y))
curve(pt(xseq, df = n1+n2-2), xname="xseq", add =TRUE, col = 2)
```
