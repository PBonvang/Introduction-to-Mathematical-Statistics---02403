The outputs are from a normal distribution.  
`pnorm(2)`: $P(X\le 2)$ in the standard normal distribution  
`pnorm(2,1,1)`: $P(X\le 2)$ in the normal distribution: $\mathcal{N}(1,1)$  
`pnorm(2,1,2)`: $P(X\le 2)$ in the normal distribution: $\mathcal{N}(1,2^2)$

```r
df1 <- data.frame(x = seq(-4, 4, length.out = 1000)) %>%
    mutate(y = dnorm(x))

ggplot(df1, aes(x, y)) + 
    geom_area(fill = "sky blue") + 
    gghighlight(x <= 2)
```

## b)
We wish to determine $P(90 <L - L_{beam} < 110)$  

1. We know that $V(L - L_{beam}) = 9 + \sigma_{beam}^2$
2. Now we can transform to standard normal distribution, where mu is 100 as 3000 - 2900 = 100.
    $$Z=\frac{(L - L_{beam})-100}{\sqrt{9 + \sigma_{beam}^2}}$$
    $$
    0.99=P\left(90<L-L_{\text {beam }}<110\right)=P\left(\frac{-10}{\sqrt{9+\sigma_{\text {beam }}^{2}}}<Z<\frac{10}{\sqrt{9+\sigma_{\text {beam }}^{2}}}\right)
    $$
    The boundaries -10 and 10 is the given error boundaries 90 and 110 minus the mean 100. 
3. With the boundaries determined for Z, using `qnorm(0.005)`$\approx -2.58$ we find the value required for the error to be below 99 percent, as each tail must max be 0.5 % therefore 0.005. With the required value determined we can insert it in the equation from earlier and determine the variance:
    $$\begin{align*}
    -2.58 &= \frac{-10}{\sqrt{9 + \sigma_{beam}^2}}\\
    \sigma_{\text {beam }}&=\sqrt{\left(\frac{10}{2.576}\right)^{2}-9}=2.464
    \end{align*}$$
