# E3.7

```r
# a)
Pulse_end <- c(173,175,174,183,181,180,170,182,188,
178,181,183,185)
Pulse_1min <- c(120,115,122,123,125,140,108,133,134,
121,130,126,128)

t.test(Pulse_end,Pulse_1min, conf.level = 0.99, pair = TRUE)
```

# E3.8

```r
ymax = 2.508
ymin = 2.103
smax = 0.3373
smin = 0.2834
sd = 0.09664
n = 10

# a)
t_val = qt(0.975, df=n-1); t_val

ymax-ymin + c(-1,1)*t_val*sd/sqrt(n)
```

```r
# b)
# As it's part of the confidence interval there's little evidence.
# The null hypothesis is: H_0: mu_0 = 0.35
delta = 0.35
tobs = (ymax-ymin-delta)/(sd/sqrt(n)); tobs

pval = 2*(1-pt(abs(tobs),df=n-1));pval
# There's alot of evidence that the true difference is 0.35
```

```r
n1 = 50
n2 = 24
nf1 = 13
nf2 = 3
m1 = 6.420
m2 = 7.375
s1 = 2.205
s2 = 1.813

# a)
tobs = (m1-m2)/sqrt(s1^2/n1 + s2^2/n2)
v = (s1^2/n1 + s2^2/n2)^2/((s1^2/n1)^2/(n1-1) + (s2^2/n2)^2/(n2-1))
v

pval = 2*(1-pt(abs(tobs), df = v)); pval
# Theres weak evidence agains H_0
```
